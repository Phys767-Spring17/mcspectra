```python
# Sample random data

import numpy as np
import matplotlib.pyplot as plt

y_array1 = np.random.randn(100)*10
y_array2 = np.random.rand(100)*10
x_array = range(100)
unc1 = np.random.randn(100)*5
unc2 = np.random.randn(100)*5
n=500
plt.plot(x_array,y_array1, color='b')
plt.plot(x_array, y_array2, color='r')
plt.fill_between(x_array, y_array1+unc1, y_array1-unc1, color='b', alpha=0.3)
plt.fill_between(x_array, y_array2+unc2, y_array2-unc2, color='r', alpha=0.3)
plt.show()
```

```python
# Function that does n iterations of RMS calculation

import numpy as np
import matplotlib.pyplot as plt

def mc_rms (y_array1, y_array2, unc1, unc2, n):

    RMS=[]

    for w in range(n):
        y_array1_rand = y_array1 + (np.random.randn(len(y_array1)) * unc1)
        y_array2_rand = y_array2 + (np.random.randn(len(y_array2)) * unc2)
        diff = y_array1_rand - y_array2_rand
        rms = (np.sum(diff**2))/(len(diff))**(0.5)
        RMS.append(rms)
       
    avg = np.mean(RMS)
    error = np.std(RMS)
    print(avg, '+/-', error)
    
    RMS.sort()
    plt.hist(RMS, bins=50)
    plt.show()

    
#     print(RMS)
```

```python
# Running the code with random sample data
mc_rms(y_array1, y_array2, unc1, unc2, n)
```
